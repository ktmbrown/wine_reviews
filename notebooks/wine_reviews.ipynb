```python
# libraries
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# getting data and writing dataframe
df = pd.read_csv('../data/raw/winemag-data-130k-v2.csv')
```

```python
# getting rid of nan's
df.dropna(inplace=True)
```

```python
# scatter plot points vs price
plt.scatter(x=df.points, y=df.price)
plt.show()
```

```python
# boxplot to check for outliers
sns.boxplot(x=df['price'])
```

```python
# getting sd and mean of price
price_sd =  np.std(df['price'])
price_mean = np.mean(df['price'])
```

```python
# checking for values outside 3 stds of the mean
df = df[df['price'] < price_mean + (3 * price_mean)]
```

```python
# scatter plot points vs price after removing outlier
plt.scatter(x=df.points, y=df.price)
plt.show()
```

```python
# boxplot to check for outliers after removing outlier
sns.boxplot(x=df['price'])
```

```python
# linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x=df.points, y=df.price)
print("r-squared:", r_value**2)
print('p_value: %6.7f' % p_value)
```

```python
# plotting linear model
plt.plot(df.points, df.price, 'o', label='original data')
plt.plot(df.points, intercept + slope*(df.points), 'r', label='fitted line')
plt.legend()
plt.show()
```
